
#**Jaynes–Cummings model (Interpretation of Floquet States)**

How to interpret quasi-energies and Floquet states of periodically driven
systems? In order to gain some intuition it is instructive to analyze an example
from the quantum optics.
Let us consider a two-level atom described by the Hamiltonian of the form
$\hat{H}_A = \omega_0(|e\rangle\langle e|-|g\rangle\langle g|)/2$ where $\omega_0$ is the energy difference between the excited and ground state levels (remember that $\hbar=1$) and $|g\rangle$ and $|e\rangle$ are the
ground and excited eigenstates of the atom, respectively. The atom is at rest
in the electromagnetic cavity and we focus on interaction of the atom with a
single electromagnetic mode of the cavity corresponding to the photon energy
$\omega\approx\omega_0$. The Hamiltonian of the entire system, i.e. the atom and the quantum electromagnetic field, in the rotating wave approximation, reads

$$
\hat{H}= \frac{\omega_0}{2}(|e\rangle\langle e|-|g\rangle\langle g|)+\omega\hat{a}^{\dagger}\hat{a}+\lambda\left(|e\rangle\langle g|\hat{a}+|g\rangle\langle e|\hat{a}^{\dagger}\right), \hspace{1cm} (4.4)
$$

with $λ = -\langle e|\vec{d}\cdot\vec{e}_0|g\rangle$ which is assumed to be real-valued, where d is the atomic electric dipole moment and the vector e0 describes the mode of the electromagnetic field, especially its polarization. The second term in the Hamiltonian (4.4) corresponds to the energy of the electromagnetic field $(\hat{a}^{\dagger}\hat{a}$ is the photon number operator) while the third term describes interaction between the field and the atom where $\hat{a}$ is the photon annihilation operator. The system we consider is the well-known Jaynes–Cummings model.
The energy of the entire system is conserved. It is easy to find eigenstates
of (4.4) (which in the quantum optics language are called dressed states of
an atom) because the Hamiltonian matrix can be written in a block-diagonal
form. That is, the Hilbert space can be divided into two-dimensional subspaces $\{|g,n + 1\rangle, |e,n\rangle\}$, where $|n\rangle$ denotes the state of the electromagnetic field with exactly n photons, which are not coupled to each other by the Hamiltonian. The two-dimensional blocks of the Hamiltonian matrix read

$$H^{(n)}=\begin{pmatrix} \frac{1}{2}\omega_0+n\omega & \lambda\sqrt{n+1} \\ \lambda\sqrt{n+1} & -\frac{1}{2}\omega_0+(n+1)\omega\end{pmatrix}, \hspace{1cm} (4.5) $$

and their diagonalization leads to the dressed atom spectrum

$$
E_{\pm}^{(n)}=\left(n+\frac{1}{2}\right)\omega\pm\frac{1}{2}\sqrt{(\omega_0-\omega)^2+4\lambda^2(n+1)} \hspace{1cm} (4.6)
$$

and the corresponding dressed states of an atom which are superposition of
$|g,n + 1\rangle$ and $|e,n\rangle$. For weak interaction between the atom and the field (i.e., $|λ|\sqrt{n+1}<<\omega_0$) and for $|\omega_0-\omega|<<\omega$, the spectrum consists of pairs of energy levels which are well separated by the photon energy $\omega$. For $n>>1$ the spectrum can be considered as locally periodic with the period $\omega$ because $E_+^{(n)}-E_-^{(n)}$ changes little when we change $n$ by $\Delta n<< n$.

Now assume that we treat the electromagnetic field as a classical wave.
Then, its interaction with a two-level atom is described by $-\vec{d}\cdot \vec{E}_0\cos{ωt}$ where $\vec{E}_0$ is the electric field vector. Because the Hamiltonian of the system is time periodic we can define the Floquet Hamiltonian which in the rotating wave approximation reads

$$
\hat{H}_F= \frac{\omega_0}{2}(|e\rangle\langle e|-|g\rangle\langle g|)+\frac{V}{2}\left(|e\rangle\langle g|e^{-i\omega t}+|g\rangle\langle e|e^{i\omega t}\right)-i\frac{\partial}{\partial t}, \hspace{1cm} (4.7)
$$

where $V=-\langle e|\vec{d}\cdot\vec{E}_0|g\rangle$. Floquet states of the system are time periodic, $|u(t+2\pi/\omega)\rangle=|u(t)\rangle$, and can be expanded in Fourier series

$$
|u(t)\rangle=\sum_k c_ke^{ik\omega t}|u_k\rangle \hspace{1cm} (4.8)
$$

When we define the Fourier basis as $\langle t|k\rangle = e^{ik\omega t}$ we can see that the matrix of the Floquet Hamiltonian in the Hilbert space extended by the time degree of freedom decomposes into two-dimensional uncoupled blocks corresponding to the subspaces $\{|g,k + 1\rangle,|e,k\rangle}, i.e

$$H_F^{(k)}=\begin{pmatrix} \frac{1}{2}\omega_0+k\omega & \frac{V}{2} \\ \frac{V}{2} & -\frac{1}{2}\omega_0+(k+1)\omega\end{pmatrix}, \hspace{1cm} (4.9) $$

Note that the calculations of the matrix elements of $\hat{H}_F$ involve integration in time over the period of the electromagnetic wave, e.g.

$$
\langle k',g|\hat{H}_F|g,k\rangle=\frac{\omega}{2\pi}\int_0^{2\pi/\omega}dt\langle g|e^{-ik'\omega t}\hat{H}_Fe^{ik\omega t}|g\rangle \hspace{1cm} (4.10)
$$

The corresponding quasi-energy spectrum

$$
E_{\pm}^{(k)}=\left(k+\frac{1}{2}\right)\omega\pm\frac{1}{2}\sqrt{(\omega_0-\omega)^2+V^2} \hspace{1cm} (4.11)
$$

is periodic with the period $\omega$, as it should, and it reproduces the dressed atom
spectrum (4.6) if there is a large number of photons $(n>>1)$, i.e. for $V\approx 2\lambda\sqrt{n}$.

Thus, in the regime of $n>>1$ the quasi-energy spectrum of the Floquet
formalism, which treats the electromagnetic field as a classical wave, matches
the dressed atom spectrum obtained in the full quantum theory of the
electromagnetic field. In other words, if the quantum field is strong, its
modification due to the interaction with an atom can be neglected and the
Floquet approach allows us to obtain good approximation for the eigenstates
of the entire atom-field system. Thus, Floquet states can be considered as
an approximation of true eigenstates of an entire system when back-action
on a subsystem that provides a time-periodic driving (here the quantum
electromagnetic field) can be neglected.

```python
#definición de los parámetros
omega_0 = 1
h_bar =  1
omega= omega_0/2

#definición del hamiltoniano dependiente del tiempo

T = (2*np.pi)/omega
psi0 = qtp.basis(2,0)
omega_vec = np.linspace(-5, 5, 75) * omega
q_energies = np.zeros((len(omega_vec), 2))
q_energies_analytic = np.zeros((len(omega_vec), 2))  # for storing analytic quasienergies

args = {'w': omega}

for idx, V in enumerate(omega_vec):
  H0 = (1/2)*h_bar*omega_0*qtp.sigmaz()
  H1 = (1/2)*V*qtp.Qobj([[0,1],[0,0]])
  H2 = (1/2)*V*qtp.Qobj([[0,0],[1,0]])
  H = [H0, [H1, lambda t, args: np.exp(-1j*args['w']*t)],[H2, lambda t, args: np.exp(1j*args['w']*t)]]
  f_modes_0, f_energies = qtp.floquet_modes(H, T, args,False)
  q_energies[idx,:] = f_energies

  # Calculate analytic quasienergies
  E_plus = -0.5*omega+0.5 * np.sqrt((omega_0 - omega)**2 + V**2)
  E_minus = 0.5*omega-0.5 * np.sqrt((omega_0 - omega)**2 + V**2)
  q_energies_analytic[idx, :] = [E_minus, E_plus]

plt.figure()
plt.figure(figsize=(11,7))
plt.plot(omega_vec, q_energies[:,0], 'bo', omega_vec, q_energies[:,1], 'ro')
plt.plot(omega_vec, q_energies_analytic[:,0], 'r--', omega_vec, q_energies_analytic[:,1], 'b--')
plt.xlabel(r'$V$')
plt.ylabel(r'Quasienergy')
plt.title(r'Floquet quasienergies')
plt.legend(['Numerical, minus state', 'Numerical, plus state', 'Analytical, minus state', 'Analytical, plus state'])
plt.show()
```

```python
                                                             # decompose the inital state in the floquet modes
omega_0 = 1
h_bar =  1
omega= omega_0/2
V=0.5

#definición del hamiltoniano dependiente del tiempo

T = (2*np.pi)/omega
tlist = np.linspace(0.0, 2*T, 100)
psi0 = qtp.basis(2,0)
args = {'w': omega}

H0 = (1/2)*h_bar*omega_0*qtp.sigmaz()
H1 = (1/2)*V*qtp.Qobj([[0,1],[0,0]])
H2 = (1/2)*V*qtp.Qobj([[0,0],[1,0]])
H = [H0, [H1, lambda t, args: np.exp(-1j*args['w']*t)],[H2, lambda t, args: np.exp(1j*args['w']*t)]]

f_modes_0, f_energies = qtp.floquet_modes(H, T, args,True)

f_coeff = qtp.floquet_state_decomposition(f_modes_0, f_energies, psi0)


# calculate the wavefunctions using the from the floquet modes
p_ex = np.zeros(len(tlist))
for n, t in enumerate(tlist):
    psi_t = qtp.floquet_wavefunction_t(f_modes_0, f_energies, f_coeff, t, H, T, args)
    p_ex[n] = qtp.expect(qtp.num(2), psi_t)

# For reference: calculate the same thing with mesolve
p_ex_ref = qtp.mesolve(H, psi0, tlist, [], [qtp.num(2)],args).expect[0]

# plot the results
plt.figure(figsize=(19,5))
plt.plot(tlist, np.real(p_ex),     'ro')
plt.plot(tlist, np.real(p_ex_ref), 'r')
plt.xlabel('Time')
plt.ylabel('Occupation probability of ground state')
plt.legend(("Floquet","Lindblad"))
plt.show()
```

```python
plt.figure(figsize=(19,5))
plt.plot(tlist, 1-np.real(p_ex), 'bo')
plt.plot(tlist, 1-np.real(p_ex_ref), 'b')
plt.xlabel('Time')
plt.ylabel('Occupation probability excited state')
plt.legend(("Floquet", "Lindblad"))
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(tlist, np.real(p_ex),'ro',tlist, 1-np.real(p_ex), 'bo')
plt.plot(tlist, np.real(p_ex_ref), 'r',tlist, 1-np.real(p_ex_ref), 'b')
plt.xlabel('Time')
plt.ylabel('Occupation probability')
plt.legend(("Floquet $P_1$","Floquet $P_0$","Lindblad $P_1$","Lindblad $P_0$"))
plt.show()
```
